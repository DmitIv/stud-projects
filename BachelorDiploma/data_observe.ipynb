```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

from data import get_data_paths
from fastai.vision import open_image, open_mask, plt

import warnings
warnings.filterwarnings("ignore")

from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
data_paths = get_data_paths("train")
image = open_image(data_paths["images"][200])
label = open_mask(data_paths["labels"][200])

print(label.shape)
print(label.data)

_,axs = plt.subplots(1,3, figsize=(12,6))
image.show(ax=axs[0], title='no mask')
image.show(ax=axs[1], y=label, title='masked')
label.show(ax=axs[2], title='mask only', alpha=1.)
```
